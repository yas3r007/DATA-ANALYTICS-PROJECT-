```python
#select * from movies where rankscore>9;"
movies[movies["rankscore"]>9]
```

```python
"select * from movies where rankscore>9 and year>2000;"

movies[(movies["rankscore"]>9) & (movies["year"]>2000)]
```

```python
#select * from movies where rankscore>9 or year>2000;"

movies[movies["rankscore"]>9 | (movies["year"]>2000)]
```

```python
#"select  * from movies where not year<=2000;"

movies[movies['year']>2000]
```

```python
#select * from movies where rankscore is null;"

movies[movies["rankscore"].isnull()]
```

```python
#select * from movies where rankscore is not null;"

movies[~movies["rankscore"].isnull()]
```

```python
#select * from movies where year in (2000,2002,2004);

movies[movies["year"].isin([2000,2002,2004])]
```

```python
#"select * from movies where year not in(2000,2002,2004);"

movies[~movies["year"].isin([2000,2002,2004])]
```

```python
#select * from movies where year between 1999 and 2003;"

movies[(movies["year"]>=1999) & (movies["year"]<=2003)]
```
